```python
num_cols = ['Daily Steps', 'Sleep Duration', 'Heart Rate']
fig, axes = plt.subplots(1, len(num_cols), figsize=(6 * len(num_cols), 4))

for ax, col in zip(axes, num_cols):
    sns.histplot(df[col], kde=True, ax=ax)
    ax.set_title(f'Распределение {col}')
    ax.set_xlabel(col)
    ax.set_ylabel('Частота')

plt.tight_layout()
plt.show()
```

Все три графика отличаются от нормального распределения. Первый график имеет несколько пиков. Во втором графике концентрация в районе 6.5 часов, 7.3 и 8 часов. В третьем графике у большинства сердцебиение до 75

```python
sns.histplot(data=df, x="Sleep Duration", hue="Gender", multiple="stack", bins=13)
plt.title("Распределение Продолжительность сна по полу")
plt.show()
```


* Большинство женского пола спят в районе 6.5-7.5 и от 8 часов
* Мужской пол в основном спит от 7 до 8 часов



```python
sns.kdeplot(data=df, x="Stress Level", hue="Gender", fill=True)
plt.title("Уровень стресса по полу")
plt.show()
```

У мужского пола уровень стресса больше, чем у женского

```python
sns.countplot(x='Blood Pressure', data=df)
plt.title('Распределение Blood Pressure')
plt.xlabel('Blood Pressure')
plt.ylabel('Количество')
plt.show()

sns.countplot(x='BMI Category', data=df)
plt.title('Распределение BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Количество')
plt.show()

sns.countplot(x='Stress Level', data=df)
plt.title('Распределение Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Количество')
plt.show()
```

В Категории ИМТ человека концентрируется в normal и overweight


* В Категории ИМТ человека концентрируется в normal и overweight
* В уровне стресса практически у всех одинаковое количество от 3 до 8




```python
sns.boxplot(x=df['Heart Rate'])
plt.title('Heart Rate')
plt.xlabel('Heart Rate')
plt.show()

sns.boxplot(x=df['Sleep Duration'])
plt.title('Sleep Duration')
plt.xlabel('Sleep Duration')
plt.show()
```

Столбец Hear Rate имеет выбросы после 80

```python
# Корреляционная матрица
numeric_df = df.select_dtypes(include=['int64', 'float64'])
correlation_matrix = numeric_df.corr()

# Визуализация тепловой карты
sns.heatmap(correlation_matrix, annot=True, cmap='coolwarm')
plt.title('Корреляционная матрица')
plt.show()
```

сильная корреляция считаем abs(corr_coef) >= 0.7
*  Sleep Duration имеет сильную корреляцию с Quality of sleep, а слабую с Stress Level
*   Physical Activity имеет сильную корреляцию с Daily steps
*   Stress Level имеет слабую корреляцую с Sleep Duration
