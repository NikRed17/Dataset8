```python
sns.histplot(df['Daily Steps'], kde=True)
plt.title('Распределение Daily Steps')
plt.xlabel('Daily Steps')
plt.ylabel('Кол-во')
plt.show()

sns.histplot(df['Sleep Duration'], kde=True)
plt.title('Распределение Sleep Duration')
plt.xlabel('Sleep Duration')
plt.ylabel('Кол-во')
plt.show()

sns.histplot(df['Heart Rate'], kde=True)
plt.title('Распределение Heart Rate')
plt.xlabel('Heart Rate')
plt.ylabel('Кол-во')
plt.show()
```

Все три графика отличаются от нормального распредеделния. Первый график имеет несколько пиков. Во втором графике концентрация в районе 6.5 часов, 7.3 и 8 часов. В третьем графике у большинства сердцебиение до 75

```python
sns.countplot(x='BMI Category', data=df)
plt.title('Распределение BMI Category')
plt.xlabel('BMI Category')
plt.ylabel('Количество')
plt.show()

sns.countplot(x='Stress Level', data=df)
plt.title('Распределение Stress Level')
plt.xlabel('Stress Level')
plt.ylabel('Количество')

```

```python
sns.boxplot(x=df['Heart Rate'])
plt.title('Heart Rate')
plt.xlabel('Heart Rate')
plt.show()
```

сильная корреляция считаем abs(corr_coef) >= 0.7
*  Sleep Duration имеет сильную корреляцию с Quality of sleep, а слабую с Stress Level
*   Physical Activity имеет сильную корреляцию с Daily steps
*   Stress Level имеет слабую корреляцую с Sleep Duration
